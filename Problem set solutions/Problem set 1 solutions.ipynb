```python
import numpy as np
from scipy.stats import norm

def Black_Scholes(S, sigma, K, T, r):
    omega = (r*T + sigma**2*T/2 - np.log(K/S)) / (sigma*np.sqrt(T))
    return S*norm.cdf(omega) - K*np.exp(-r*T)*norm.cdf(omega - sigma*np.sqrt(T))

sigma = 0.3
K = 50
T = 2
r = 0.1
```

```python
Black_Scholes(20, sigma, K, T, r)
```

```python
Black_Scholes(50, sigma, K, T, r)
```

```python
Black_Scholes(80, sigma, K, T, r)
```

```python
from matplotlib import pyplot as plt
%matplotlib inline

S = np.linspace(0.01, 80, 1000)

def C(t):
    return Black_Scholes(S, sigma, K, T - t, r)

plt.figure(figsize = (10, 5))

for t in (0, 1, 1.5, 1.9):
    plt.plot(S, C(t), label = '$C(S,{})$'.format(t))

def payoff(S, K):
    if S > K:
        return S - K
    return 0

plt.plot(S, [payoff(S0, K) for S0 in S], label = 'Payoff = $C(S,{})$'.format(T))

plt.xlabel('$S$')
plt.ylabel('$C(S)$', rotation = 'horizontal', labelpad = 16)
plt.legend()

plt.savefig('Black-Scholes.png', dpi=600)
plt.show()
```
